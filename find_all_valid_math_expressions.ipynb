### Find All Valid Math Expressions
You are attempting to solve a Coding Contract. You have 10 tries remaining, after which the contract will self-destruct.


You are given the following string which contains only digits between 0 and 9:

295626

You are also given a target number of -72. Return all possible ways you can add the +(add), -(subtract), and *(multiply) operators to the string such that it evaluates to the target number. (Normal order of operations applies.)

The provided answer should be an array of strings containing the valid expressions. The data provided by this problem is an array with two elements. The first element is the string of digits, while the second element is the target number:

["295626", -72]

NOTE: The order of evaluation expects script operator precedence NOTE: Numbers in the expression cannot have leading 0's. In other words, "1+01" is not a valid expression Examples:

Input: digits = "123", target = 6
Output: [1+2+3, 1*2*3]

Input: digits = "105", target = 5
Output: [1*0+5, 10-5]

```python
def find_all_valid_math_expressions(digits, target):
    def backtrack(index, path, value, last):
        # If we reach the end of the string, check if the current expression evaluates to the target
        if index == len(digits):
            if value == target:
                result.append(path)
            return

        for i in range(index, len(digits)):
            # Avoid numbers with leading zero
            if i != index and digits[index] == '0':
                break

            # Take the current number
            current_number = int(digits[index:i+1])

            # If at the start of the string, just pass the number to the next call
            if index == 0:
                backtrack(i + 1, str(current_number), current_number, current_number)
            else:
                # Addition
                backtrack(i + 1, path + '+' + str(current_number), value + current_number, current_number)
                # Subtraction
                backtrack(i + 1, path + '-' + str(current_number), value - current_number, -current_number)
                # Multiplication
                backtrack(i + 1, path + '*' + str(current_number), value - last + last * current_number, last * current_number)

    result = []
    backtrack(0, '', 0, 0)
    return result
```
